# SDKs y frameworks para agentes de IA

## El ecosistema de herramientas para construir agentes

El desarrollo de agentes de IA ha dado lugar a un ecosistema diverso de **SDKs y frameworks** que simplifican la construcción de sistemas autónomos. En los últimos años, tanto los grandes proveedores de modelos como la comunidad open source han lanzado herramientas especializadas para este propósito.

El panorama actual se divide en dos grandes categorías: los **SDKs de proveedores**, desarrollados por las empresas que ofrecen los modelos de lenguaje, y los **frameworks independientes**, que permiten trabajar con múltiples proveedores de forma agnóstica.

> La elección del framework adecuado depende del ecosistema tecnológico de la organización, los modelos que se utilizarán y las necesidades específicas en términos de **colaboración multiagente**, **observabilidad** y **control sobre el comportamiento** del agente.

## LangChain y LangGraph

**LangChain** es el framework más popular y maduro para el desarrollo de aplicaciones basadas en modelos de lenguaje. Creado en 2022, ha evolucionado significativamente hasta convertirse en un estándar de facto en la industria, con una comunidad activa de miles de desarrolladores y una extensa documentación.

Con el lanzamiento de la **versión 1.0**, LangChain introdujo una nueva abstracción central para agentes: la función `create_agent`. Esta API simplificada representa un cambio importante respecto a versiones anteriores, ofreciendo una interfaz más coherente y fácil de usar para la mayoría de casos de uso.

### Características principales de LangChain

* **Agnóstico al proveedor**: Soporta modelos de OpenAI, Anthropic, Google, Mistral, Cohere y modelos locales a través de Ollama, entre otros. Cambiar de proveedor requiere modificar únicamente la configuración del modelo.

* **Sistema de herramientas flexible**: Las herramientas se definen como funciones de Python con decoradores, permitiendo al agente interactuar con APIs externas, bases de datos, sistemas de archivos o cualquier servicio que el desarrollador necesite integrar.

* **Middleware extensible**: Permite insertar lógica personalizada en diferentes puntos del ciclo del agente, habilitando funcionalidades como detección de datos sensibles, compresión de historial, aprobación humana o logging avanzado.

* **Salidas estructuradas**: El agente puede generar respuestas que se ajusten a un esquema predefinido, garantizando que los datos producidos tengan el formato esperado por los sistemas consumidores.

* **Observabilidad con LangSmith**: Ofrece integración nativa con LangSmith, una plataforma de observabilidad que permite trazar, evaluar y depurar el comportamiento de los agentes en desarrollo y producción.

### LangGraph: control a bajo nivel

Para escenarios que requieren **control preciso** sobre el flujo de ejecución, LangChain ofrece **LangGraph**, una librería complementaria que permite definir agentes como grafos de estados. Mientras que `create_agent` proporciona una abstracción de alto nivel lista para usar, LangGraph expone una **Graph API** que permite:

* Definir nodos y aristas del flujo de ejecución de forma explícita
* Implementar lógica de bifurcación y ciclos condicionales
* Gestionar estados persistentes con checkpointing
* Controlar la ejecución paso a paso para debugging

> LangChain con `create_agent` es ideal para empezar rápidamente, mientras que **LangGraph** ofrece el control necesario para arquitecturas complejas donde el flujo del agente debe ser completamente personalizado.

Este curso utiliza LangChain como framework principal debido a su madurez, versatilidad y la nueva API `create_agent` que simplifica significativamente la curva de aprendizaje.

## OpenAI Agents SDK

**OpenAI** lanzó su **Agents SDK** en marzo de 2025 como una herramienta de código abierto para construir aplicaciones con agentes. Este SDK está optimizado para trabajar con los modelos de OpenAI y proporciona una experiencia integrada con su ecosistema de APIs.

### Características principales

* **Agentes declarativos**: Los agentes se configuran mediante un modelo declarativo que especifica instrucciones, herramientas disponibles y comportamiento esperado, reduciendo la cantidad de código necesario.

* **Handoffs entre agentes**: El SDK permite que un agente transfiera el control a otro agente especializado de forma fluida, manteniendo el contexto de la conversación. Este patrón es útil para sistemas donde diferentes agentes manejan diferentes dominios.

* **Guardrails configurables**: Sistema de validación que permite definir restricciones sobre las entradas que el agente acepta y las salidas que genera, añadiendo una capa de seguridad.

* **Trazabilidad integrada**: Incluye herramientas de visualización que permiten inspeccionar cada decisión y llamada a herramienta, facilitando la depuración y optimización.

* **Integración con APIs de OpenAI**: Acceso nativo a funcionalidades como file search, code interpreter y la API de Assistants.

```mermaid

flowchart LR
    subgraph "OpenAI Agents SDK"
        A1[Agente Triaje] -->|Handoff| A2[Agente Ventas]
        A1 -->|Handoff| A3[Agente Soporte]
        A2 --> R[Respuesta]
        A3 --> R
    end
    
    U[Usuario] --> A1
```

El SDK de OpenAI es adecuado para proyectos que trabajan exclusivamente con modelos de OpenAI y buscan una integración profunda con sus servicios específicos.

## Google Agent Development Kit

El **Agent Development Kit (ADK)** de Google, anunciado en abril de 2025, es un framework modular diseñado para integrarse con el ecosistema de **Google Cloud Platform**, incluyendo los modelos **Gemini** y la plataforma **Vertex AI**.

### Características principales

* **Integración nativa con Google Cloud**: Optimizado para trabajar con servicios como BigQuery, Cloud Storage, Firestore y las APIs de Gemini. Los agentes pueden acceder a estos servicios con configuración mínima.

* **Composiciones jerárquicas**: Soporta arquitecturas donde un agente supervisor coordina a múltiples agentes especializados, delegando subtareas y consolidando resultados.

* **Desarrollo conciso**: Las abstracciones de alto nivel permiten construir agentes funcionales con relativamente pocas líneas de código.

* **Herramientas predefinidas**: Incluye un catálogo de herramientas listas para usar que cubren casos comunes como búsqueda, generación de código y acceso a datos.

* **Despliegue en Vertex AI**: Los agentes pueden desplegarse directamente en Vertex AI Agent Builder para escalado y gestión en producción.

ADK es especialmente adecuado para organizaciones que ya utilizan Google Cloud Platform y buscan una integración fluida con sus servicios de datos e infraestructura existentes.

## Anthropic Claude Agent SDK

**Anthropic** ofrece el **Claude Agent SDK**, un kit de desarrollo que permite construir agentes de IA que operan de forma autónoma: leen archivos, ejecutan comandos, buscan en la web, editan código y más. El SDK proporciona las mismas herramientas, bucle de agente y gestión de contexto que alimentan a Claude Code, pero programables desde Python y TypeScript.

### Características principales

* **Herramientas integradas**: El SDK incluye herramientas listas para usar como Read, Write, Edit, Bash, Glob, Grep, WebSearch y WebFetch, permitiendo que el agente comience a trabajar inmediatamente sin implementar la ejecución de herramientas manualmente.

* **Hooks para el ciclo de vida**: Permite ejecutar código personalizado en puntos clave del ciclo del agente mediante callbacks, posibilitando validar, registrar, bloquear o transformar el comportamiento del agente.

* **Subagentes**: Soporta la creación de agentes especializados que manejan subtareas específicas. El agente principal delega trabajo y los subagentes reportan resultados.

* **Integración con MCP**: Compatible con el Model Context Protocol para conectar con sistemas externos como bases de datos, navegadores, APIs y cientos de servidores MCP disponibles.

* **Gestión de sesiones**: Permite mantener el contexto entre múltiples consultas, habilitando conversaciones continuas donde el agente recuerda interacciones previas.

```mermaid

flowchart TD
    subgraph "Claude Agent SDK"
        AGENT[Agente Principal]
        
        subgraph Tools["Herramientas Built-in"]
            T1[Read / Write / Edit]
            T2[Bash / Glob / Grep]
            T3[WebSearch / WebFetch]
        end
        
        subgraph Sub["Subagentes"]
            S1[Code Reviewer]
            S2[Researcher]
        end
    end
    
    AGENT --> Tools
    AGENT --> Sub
    AGENT <-->|MCP| EXT[Servidores Externos]
```

> El Claude Agent SDK destaca por incluir **herramientas de ejecución listas para usar**, eliminando la necesidad de implementar la lógica de ejecución de herramientas que otros SDKs requieren.

Anthropic también desarrolla el **Model Context Protocol (MCP)**, un estándar abierto que define cómo los modelos de lenguaje pueden comunicarse con herramientas y fuentes de datos externas. MCP ha sido adoptado por otros proveedores como OpenAI y Google, convirtiéndose en un estándar emergente para la interoperabilidad de herramientas.

## CrewAI

**CrewAI**, lanzado a finales de 2023, es un framework de código abierto diseñado específicamente para la creación de sistemas **multiagente**. Su metáfora central es la de un equipo o "crew" de agentes especializados que colaboran para completar tareas complejas.

### Características principales

* **Agentes con personalidad**: Cada agente se define con un rol, objetivos específicos, una "historia de fondo" y un conjunto de herramientas. Esta configuración guía cómo el agente aborda las tareas asignadas.

* **Tareas estructuradas**: El trabajo se organiza en tareas con descripciones claras, resultados esperados y dependencias entre ellas, permitiendo flujos de trabajo complejos.

* **Procesos de colaboración**: Define cómo los agentes trabajan juntos: de forma secuencial, jerárquica o mediante consenso.

* **Agnóstico al proveedor**: Soporta modelos de OpenAI, Anthropic, modelos locales vía Ollama y otros proveedores.

* **Herramientas integradas**: Incluye herramientas predefinidas para búsqueda web, scraping, ejecución de código y llamadas a APIs.

```mermaid

flowchart LR
    subgraph "CrewAI"
        subgraph Crew["Equipo"]
            A1[Investigador]
            A2[Redactor]
            A3[Editor]
        end
    end
    
    T[Tarea] --> A1
    A1 -->|Investigación| A2
    A2 -->|Borrador| A3
    A3 --> R[Resultado Final]
```

CrewAI es adecuado para proyectos que requieren múltiples agentes colaborando en flujos estructurados, como pipelines de investigación, generación de contenido o análisis de datos que involucran diferentes perspectivas o especialidades.

## Comparativa de frameworks

| Framework | Enfoque | Multiagente | Mejor para |
|-----------|---------|-------------|------------|
| **LangChain** | Flexibilidad y extensibilidad | Sí (via LangGraph) | Proyectos que requieren control fino y agnósticos al proveedor |
| **LangGraph** | Control a bajo nivel | Sí (grafos de estados) | Flujos complejos con lógica de ejecución personalizada |
| **OpenAI SDK** | Integración con OpenAI | Sí (Handoffs) | Proyectos centrados en modelos OpenAI |
| **Google ADK** | Ecosistema Google Cloud | Sí (jerárquico) | Organizaciones en Google Cloud Platform |
| **Claude Agent SDK** | Herramientas built-in | Sí (Subagentes) | Agentes que necesitan ejecutar código, editar archivos o buscar en web |
| **CrewAI** | Equipos de agentes | Sí (nativo) | Flujos colaborativos con roles especializados |

En este curso utilizaremos **LangChain** con su función `create_agent` como framework principal. Esta elección se basa en su equilibrio entre flexibilidad, documentación extensa y la API simplificada de la versión 1.0. Los conceptos y patrones aprendidos son transferibles a otros frameworks, ya que todos implementan variaciones del mismo patrón fundamental de agentes.
