# Fundamentos de los agentes de IA

## Qué es un agente de IA

Un **agente de inteligencia artificial** es un sistema capaz de percibir su entorno, razonar sobre la información recibida y ejecutar acciones de forma autónoma para alcanzar objetivos específicos. A diferencia de un modelo que simplemente genera texto en respuesta a una entrada, un agente toma decisiones, interactúa con herramientas externas y adapta su comportamiento según los resultados obtenidos.

La definición clásica de agente proviene del libro *Artificial Intelligence: A Modern Approach* de Stuart Russell y Peter Norvig, donde se describe como una entidad que **percibe** su entorno mediante sensores y **actúa** sobre él mediante actuadores. En el contexto de la IA generativa moderna, los sensores son las entradas de texto, imágenes o datos estructurados, mientras que los actuadores son las llamadas a APIs, la ejecución de código o la generación de respuestas.

> Un agente de IA no es simplemente un modelo de lenguaje que responde preguntas. Es un sistema que **decide qué hacer**, **ejecuta acciones** y **evalúa resultados** en un ciclo continuo hasta cumplir su objetivo.

La característica fundamental que distingue a un agente es su **autonomía**. Mientras que un LLM tradicional requiere una instrucción explícita para cada respuesta, un agente puede descomponer un objetivo complejo en subtareas, decidir qué herramientas utilizar y cuándo, y ajustar su estrategia si algo no funciona como esperaba.

```mermaid
flowchart TD
    input([input]) --> model{model}
    model -->|action| tools[tools]
    tools -->|observation| model
    model -->|finish| output([output])
```

### El ciclo percepción-razonamiento-acción

Todo agente opera mediante un **bucle iterativo** que se repite hasta alcanzar el objetivo o determinar que no es posible completar la tarea:

* **Percepción**: El agente recibe información del entorno. Esta información puede ser la consulta inicial del usuario, el resultado de una herramienta ejecutada previamente, o datos recuperados de una base de conocimiento.

* **Razonamiento**: El agente analiza la información disponible, considera sus opciones y decide el siguiente paso. Aquí es donde el LLM subyacente genera un plan o selecciona una acción.

* **Acción**: El agente ejecuta la decisión tomada. Puede invocar una herramienta externa, generar una respuesta parcial, o solicitar más información al usuario.

Este ciclo se conoce como el **agent loop** y es el patrón arquitectónico central de todos los sistemas agénticos modernos.

## Origen histórico de los agentes basados en LLM

El concepto de agente en inteligencia artificial no es nuevo. Los primeros sistemas expertos de los años 80 ya implementaban formas primitivas de razonamiento autónomo mediante reglas simbólicas. Sin embargo, la revolución actual de los agentes basados en LLM tiene un origen mucho más reciente y está directamente vinculada a varios avances de investigación publicados entre 2022 y 2023.

### Chain of Thought y el razonamiento paso a paso

En enero de 2022, investigadores de Google publicaron el paper *Chain-of-Thought Prompting Elicits Reasoning in Large Language Models*. Este trabajo demostró que los LLM pueden resolver problemas complejos de forma más efectiva cuando se les indica que **razonen paso a paso** antes de dar una respuesta final.

El descubrimiento fue significativo porque mostró que los modelos de lenguaje no solo almacenan conocimiento, sino que pueden **simular procesos de razonamiento** cuando se les estructura la tarea adecuadamente. Esta capacidad de descomponer problemas en pasos intermedios es fundamental para el funcionamiento de los agentes.

```mermaid

flowchart TD
    subgraph "Sin Chain of Thought"
        Q1[Pregunta] --> R1[Respuesta directa]
    end
    
    subgraph "Con Chain of Thought"
        Q2[Pregunta] --> P1[Paso 1: Analizar]
        P1 --> P2[Paso 2: Descomponer]
        P2 --> P3[Paso 3: Resolver]
        P3 --> R2[Respuesta razonada]
    end
```

### ReAct: la fusión de razonamiento y acción

El paper que estableció las bases teóricas de los agentes modernos fue *ReAct: Synergizing Reasoning and Acting in Language Models*, publicado en octubre de 2022 por investigadores de Princeton y Google. Este trabajo introdujo el paradigma **ReAct** (Reasoning + Acting), que combina el razonamiento en cadena de pensamiento con la capacidad de ejecutar acciones en el mundo real.

> El patrón ReAct permite que un agente alterne entre **pensar** sobre lo que debe hacer y **actuar** ejecutando herramientas, observando los resultados y ajustando su razonamiento en consecuencia.

El formato ReAct estructura la interacción del agente en tres tipos de pasos:

* **Thought**: El agente verbaliza su razonamiento interno, analizando la situación y planificando su siguiente movimiento.

* **Action**: El agente especifica qué herramienta invocar y con qué parámetros.

* **Observation**: El sistema registra el resultado de la acción ejecutada, que se incorpora al contexto para el siguiente ciclo de razonamiento.

```mermaid

sequenceDiagram
    participant U as Usuario
    participant A as Agente
    participant T as Herramienta
    
    U->>A: Dame el clima en Madrid
    Note over A: Thought: Necesito buscar<br/>información meteorológica
    A->>T: Action: weather_api(Madrid)
    T->>A: Observation: 18°C, soleado
    Note over A: Thought: Tengo la info,<br/>puedo responder
    A->>U: El clima en Madrid es 18°C y soleado
```

### Toolformer y el uso autónomo de herramientas

En febrero de 2023, Meta AI publicó *Toolformer: Language Models Can Teach Themselves to Use Tools*. Este paper demostró que los LLM pueden aprender de forma autónoma cuándo y cómo utilizar herramientas externas como calculadoras, motores de búsqueda o traductores.

La importancia de Toolformer radica en que mostró que el uso de herramientas no requiere instrucciones explícitas detalladas: un modelo suficientemente capaz puede **decidir por sí mismo** cuándo una herramienta es útil y cómo invocarla correctamente.

## Diferencias entre un LLM y un agente

Comprender la distinción entre un modelo de lenguaje y un agente es esencial para diseñar sistemas efectivos. Aunque un agente utiliza un LLM como su núcleo de razonamiento, las diferencias arquitectónicas y funcionales son sustanciales.

### El LLM como componente

Un **modelo de lenguaje grande** es una red neuronal entrenada para predecir la siguiente palabra en una secuencia de texto. Modelos como GPT, Gemini o Claude han demostrado capacidades de comprensión y generación de lenguaje natural, pero su funcionamiento es fundamentalmente **reactivo**: reciben una entrada, generan una salida, y el proceso termina.

| Característica | LLM | Agente |
|----------------|-----|--------|
| Modo de operación | Entrada-salida única | Bucle iterativo |
| Interacción con el mundo | No | Sí, mediante herramientas |
| Persistencia de estado | No nativa | Memoria de corto y largo plazo |
| Toma de decisiones | Implícita en la generación | Explícita y observable |
| Autonomía | Ninguna | Alta |

> Un LLM es como un experto que responde preguntas. Un agente es como un asistente que **hace cosas por ti**: busca información, ejecuta tareas, y te informa del resultado.

### Capacidades exclusivas de los agentes

Los agentes extienden las capacidades de un LLM en varias dimensiones:

* **Uso de herramientas**: Un agente puede invocar APIs externas, ejecutar código, consultar bases de datos o interactuar con servicios web. El LLM subyacente decide qué herramienta usar y cómo interpretar los resultados.

* **Memoria persistente**: Mientras que un LLM solo tiene acceso a la ventana de contexto actual, un agente puede mantener memoria a corto plazo de la conversación y memoria a largo plazo que persiste entre sesiones.

* **Planificación**: Un agente puede descomponer objetivos complejos en subtareas, priorizarlas y ejecutarlas en orden. Esta capacidad de planificación multinivel no existe en un LLM aislado.

* **Autocorrección**: Cuando una acción falla o produce resultados inesperados, un agente puede detectar el problema, ajustar su estrategia y reintentar con un enfoque diferente.

```mermaid

flowchart TB
    subgraph LLM["LLM Tradicional"]
        I1[Entrada] --> M1[Modelo]
        M1 --> O1[Salida]
    end
    
    subgraph Agent["Agente de IA"]
        I2[Objetivo] --> A2[LLM Core]
        A2 <--> Tools[Herramientas]
        A2 <--> Mem[Memoria]
        A2 <--> Plan[Planificador]
        A2 --> O2[Resultado + Acciones]
    end
```

## Componentes de un agente de IA

Un agente moderno está compuesto por varios subsistemas que trabajan en conjunto. Cada componente tiene una responsabilidad específica y contribuye a la capacidad general del sistema para operar de forma autónoma.

### El núcleo de razonamiento

El **LLM** actúa como el cerebro del agente. Es responsable de:

* Interpretar las instrucciones del usuario
* Analizar los resultados de las herramientas
* Decidir qué acción tomar a continuación
* Generar respuestas coherentes y útiles

La elección del modelo tiene un impacto directo en las capacidades del agente. Modelos más grandes y capaces pueden manejar tareas más complejas, pero también requieren más recursos computacionales.

### Herramientas (Tools)

Las **herramientas** son funciones o servicios externos que el agente puede invocar para interactuar con el mundo. Ejemplos comunes incluyen:

* Búsqueda web para obtener información actualizada
* Calculadoras para operaciones matemáticas precisas
* APIs de terceros para acceder a servicios específicos
* Ejecución de código para tareas programáticas
* Sistemas de archivos para leer y escribir datos

> Las herramientas transforman al agente de un sistema puramente conversacional a uno que puede **producir efectos reales** en el mundo exterior.

Cada herramienta se define con un **nombre**, una **descripción** que el LLM utiliza para decidir cuándo usarla, y un **esquema de parámetros** que especifica qué datos necesita recibir.

### Memoria

La **memoria** permite al agente mantener contexto más allá de la ventana de tokens del LLM. Se divide típicamente en:

* **Memoria a corto plazo**: Mantiene el historial de la conversación actual, incluyendo mensajes del usuario, respuestas del agente y resultados de herramientas.

* **Memoria a largo plazo**: Almacena información que persiste entre sesiones, como preferencias del usuario, hechos aprendidos o resúmenes de interacciones anteriores.

### Planificación

El módulo de **planificación** permite al agente abordar tareas complejas que requieren múltiples pasos. Algunas estrategias de planificación incluyen:

* **Planificación lineal**: Descomponer el objetivo en una secuencia ordenada de pasos.
* **Planificación jerárquica**: Crear sub-objetivos que a su vez se descomponen en tareas más pequeñas.
* **Planificación reactiva**: Ajustar el plan dinámicamente según los resultados obtenidos.

```mermaid

flowchart TD
    subgraph "Arquitectura de un Agente"
        LLM[LLM Core]
        
        subgraph Tools[Herramientas]
            T1[Búsqueda]
            T2[Calculadora]
            T3[APIs]
        end
        
        subgraph Memory[Memoria]
            M1[Corto plazo]
            M2[Largo plazo]
        end
        
        subgraph Planning[Planificación]
            P1[Descomposición]
            P2[Priorización]
        end
        
        LLM <--> Tools
        LLM <--> Memory
        LLM <--> Planning
    end
```

## Agent Engineering

La **Agent Engineering** o ingeniería de agentes es una disciplina emergente que estudia el proceso iterativo de refinar sistemas LLM no deterministas hasta convertirlos en experiencias de producción fiables. A diferencia del desarrollo de software tradicional, donde los inputs y outputs están relativamente definidos, los agentes operan en un espacio donde los usuarios pueden decir literalmente cualquier cosa y el rango de comportamientos posibles es muy amplio.

> Agent Engineering es el proceso iterativo de convertir sistemas no deterministas basados en LLM en experiencias de producción fiables. Es un ciclo continuo: **construir, probar, desplegar, observar, refinar, repetir**.

La brecha entre "funciona en mi máquina" y "funciona en producción" puede ser enorme en el desarrollo de agentes. Empresas que han logrado llevar agentes fiables a producción no siguen el playbook tradicional de desarrollo de software, sino que han desarrollado prácticas específicas para esta nueva realidad.

### Las tres disciplinas del Agent Engineering

La ingeniería de agentes combina tres conjuntos de habilidades que trabajan en conjunto:

**Product thinking**: Define el alcance y moldea el comportamiento del agente. Esto implica escribir prompts que dirigen el comportamiento del agente, que pueden llegar a ser cientos o miles de líneas. Requiere comprender profundamente el "trabajo a realizar" que el agente replica y definir evaluaciones que comprueben si el agente cumple su propósito.

**Engineering**: Construye la infraestructura que hace que los agentes estén listos para producción. Incluye escribir herramientas para que el agente las use, desarrollar UI/UX para las interacciones del agente con streaming y manejo de interrupciones, y crear runtimes robustos que manejen ejecución durable, pausas para intervención humana y gestión de memoria.

**Data science**: Mide y mejora el rendimiento del agente a lo largo del tiempo. Implica construir sistemas de evaluación, A/B testing y monitorización para medir el rendimiento y la fiabilidad del agente, así como analizar patrones de uso y errores.

### El ciclo de desarrollo de agentes

El desarrollo de agentes opera bajo un principio diferente al desarrollo de software tradicional. Para lograr un sistema de agentes fiable, **desplegar es la forma de aprender**, no lo que haces después de aprender.

```mermaid

flowchart LR
    B[Construir] --> T[Probar]
    T --> S[Desplegar]
    S --> O[Observar]
    O --> R[Refinar]
    R --> B
```

* **Construir**: Diseñar la base del agente, desde una simple llamada a LLM con herramientas hasta un sistema multiagente complejo. La arquitectura depende de cuánto workflow determinista versus agencia basada en LLM se necesita.

* **Probar**: Validar el agente contra escenarios de ejemplo para detectar problemas obvios con prompts, definiciones de herramientas y flujos de trabajo. A diferencia del software tradicional donde se pueden mapear los flujos de usuario, no es posible anticipar todas las formas en que los usuarios interactuarán con lenguaje natural.

* **Desplegar**: Una vez desplegado, se empiezan a ver inputs que no se habían considerado. Cada traza de producción muestra lo que el agente realmente necesita manejar.

* **Observar**: Trazar cada interacción para ver la conversación completa, cada herramienta invocada y el contexto exacto que informó cada decisión. Ejecutar evaluaciones sobre los datos de producción para medir la calidad del agente.

* **Refinar**: Una vez identificados patrones en lo que falla, refinar editando prompts y modificando definiciones de herramientas. Añadir los casos problemáticos al conjunto de escenarios de prueba para testing de regresión.

### Por qué los agentes son diferentes

Los agentes presentan desafíos únicos que no existen en el desarrollo de software tradicional:

* **Cada input es un edge case**: No existe un input "normal" cuando los usuarios pueden preguntar cualquier cosa en lenguaje natural. Cuando alguien escribe "hazlo mejor" o "haz lo mismo que antes pero diferente", el agente puede interpretar estas instrucciones de múltiples formas.

* **No se puede depurar de la forma tradicional**: Como gran parte de la lógica vive dentro del modelo, es necesario inspeccionar cada decisión y cada llamada a herramientas. Pequeños ajustes en prompts o configuración pueden crear cambios enormes en el comportamiento.

* **"Funcionando" no es binario**: Un agente puede tener un 99.99% de uptime mientras está completamente fuera de control. No siempre hay respuestas simples de sí o no a las preguntas importantes: ¿Está el agente tomando las decisiones correctas? ¿Usando las herramientas de la forma correcta? ¿Siguiendo la intención detrás de las instrucciones?

> La clave del éxito en Agent Engineering es dejar de intentar perfeccionar los agentes antes del lanzamiento y empezar a tratar la producción como el principal maestro.

### El Model Context Protocol

El **Model Context Protocol** (MCP) es un estándar abierto que define cómo los agentes pueden descubrir y utilizar herramientas de forma estandarizada. Introducido por Anthropic, MCP proporciona:

* Un formato común para describir herramientas y sus capacidades
* Un protocolo de comunicación entre el agente y los proveedores de herramientas
* Mecanismos de autenticación y autorización

MCP representa un paso hacia un ecosistema de herramientas interoperables, donde un mismo agente puede conectarse a múltiples fuentes de capacidades sin necesidad de integraciones personalizadas para cada una.

### Observabilidad y trazabilidad

Uno de los pilares fundamentales del Agent Engineering es la **observabilidad**. Dado que los agentes toman decisiones autónomas y ejecutan múltiples pasos, es esencial poder inspeccionar exactamente qué sucedió en cada interacción:

* **Trazas completas**: Cada conversación, herramienta invocada y contexto que informó cada decisión debe quedar registrado.
* **Evaluaciones continuas**: Sistemas que miden la calidad del agente sobre datos reales de producción.
* **Análisis de patrones**: Identificación sistemática de casos de fallo y oportunidades de mejora.

Los equipos que tienen éxito desplegando agentes fiables comparten una característica: trazan cada decisión, evalúan a escala y despliegan mejoras en días en lugar de trimestres. La iteración sistemática basada en observaciones de producción es lo que convierte un prototipo en un sistema de producción fiable.
