# Configuración del entorno para agentes de IA

El desarrollo de agentes de IA requiere un entorno de trabajo correctamente configurado con las herramientas y dependencias necesarias. Esta lección cubre la instalación de **uv** como gestor de paquetes moderno, Python 3.14, LangChain con sus integraciones y herramientas de búsqueda. La memoria (short/long term) y RAG con bases de datos se verán en lecciones posteriores.

## Instalación de uv y Python

**uv** es un gestor de paquetes para Python desarrollado por Astral. Ofrece velocidades de instalación significativamente superiores a pip y proporciona gestión integrada de versiones de Python, entornos virtuales y dependencias en una única herramienta.

```mermaid

flowchart LR
    subgraph uv["Gestor uv"]
        PY[Versiones Python]
        ENV[Entornos virtuales]
        DEP[Dependencias]
    end
    
    DEV[Desarrollador] --> uv
    uv --> PROJ[Proyecto Python]
```

### Instalación de uv

La forma recomendada de instalar uv es mediante el script de instalación oficial. En **Windows**, abre PowerShell y ejecuta:

```powershell .noeval
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

En **macOS** o **Linux**, utiliza el siguiente comando:

```bash .noeval
curl -LsSf https://astral.sh/uv/install.sh | sh
```

Tras la instalación, verifica que uv está disponible:

```bash .noeval
uv --version
```

> uv gestiona automáticamente las versiones de Python, eliminando la necesidad de instaladores externos como pyenv o el instalador oficial de Python.

### Instalación de Python 3.14

Con uv instalado, puedes descargar e instalar Python 3.14 directamente:

```bash .noeval
uv python install 3.14
```

Este comando descarga una distribución de Python precompilada y la registra en el sistema de uv. Puedes verificar las versiones instaladas con:

```bash .noeval
uv python list
```

Para establecer Python 3.14 como versión predeterminada para nuevos proyectos:

```bash .noeval
uv python install 3.14 --default
```

## Creación del proyecto

Un proyecto de agentes de IA bien estructurado facilita la gestión de dependencias y la colaboración. uv proporciona comandos para inicializar proyectos con la configuración estándar de Python.

### Inicialización con uv init

Crea un nuevo directorio para el proyecto e inicialízalo:

```bash .noeval
uv init agentes-ia
cd agentes-ia
```

Este comando genera la estructura base del proyecto:

```plaintext .noeval
agentes-ia/
├── .python-version
├── .venv/
├── pyproject.toml
├── main.py
└── README.md
```

El archivo `pyproject.toml` contiene la configuración del proyecto y las dependencias. El archivo `.python-version` especifica la versión de Python a utilizar.

> El archivo `.python-version` permite que uv active automáticamente la versión correcta de Python al entrar en el directorio del proyecto.

### Estructura recomendada

Para un proyecto de agentes de IA, la estructura de directorios puede extenderse de la siguiente manera:

```plaintext .noeval
agentes-ia/
├── .python-version
├── .venv/
├── pyproject.toml
├── .env
├── src/
│   ├── __init__.py
│   ├── agents/
│   │   └── __init__.py
│   └── tools/
│       └── __init__.py
└── README.md
```

El archivo `.env` almacenará las claves de API y configuración sensible que no debe incluirse en el control de versiones.

## Instalación de LangChain y dependencias

LangChain proporciona las abstracciones necesarias para construir agentes de IA. La versión 1.0 introdujo la función `create_agent` como la forma estándar de crear agentes, reemplazando las APIs anteriores con una interfaz más limpia y coherente.

```mermaid

flowchart TD
    subgraph "Paquetes LangChain"
        LC[langchain]
        LCO[langchain-ollama]
        LCT[langchain-tavily]
        LG[langgraph]
    end
    
    LC --> LG
    LCO --> LC
    LCT --> LC
```

### Paquete principal

Instala LangChain con uv:

```bash .noeval
uv add langchain
```

LangChain sigue una arquitectura modular donde las integraciones con proveedores específicos se distribuyen como paquetes separados. Esto permite instalar únicamente las dependencias necesarias para cada proyecto.

### Integración con Ollama

Para utilizar modelos locales a través de Ollama, instala el paquete de integración:

```bash .noeval
uv add langchain-ollama
```

Este paquete proporciona la clase `ChatOllama` para modelos de chat, que utilizaremos con el modelo gemma3:1b.

### Herramienta de búsqueda Tavily

Tavily Search es una API de búsqueda web optimizada para agentes de IA. Instala la integración:

```bash .noeval
uv add langchain-tavily
```

> Las integraciones de LangChain se distribuyen como paquetes independientes para mantener el paquete principal ligero y permitir actualizaciones independientes de cada proveedor.

### Verificación de dependencias

El archivo `pyproject.toml` debería contener las siguientes dependencias tras las instalaciones.

Para sincronizar el entorno con las dependencias declaradas:

```bash .noeval
uv sync
```

## Configuración de Ollama

**Ollama** permite ejecutar modelos de lenguaje de forma local. Proporciona una interfaz sencilla para descargar, gestionar y ejecutar modelos sin necesidad de configurar infraestructura de GPU compleja.

### Instalación de Ollama

Descarga e instala Ollama desde su sitio oficial en [ollama.com/download](https://ollama.com/download). El instalador está disponible para Windows, macOS y Linux.

En Windows, ejecuta el instalador descargado. En macOS y Linux, puedes usar el script de instalación:

```bash .noeval
curl -fsSL https://ollama.com/install.sh | sh
```

Verifica la instalación:

```bash .noeval
ollama --version
```

### Descarga de modelos

Este curso utiliza el modelo **gemma3:1b** de la familia Gemma de Google: un modelo de lenguaje compacto de 1 billón de parámetros, adecuado para tareas de razonamiento y generación de texto con recursos limitados.

Descarga el modelo:

```bash .noeval
ollama pull gemma3:1b
```

La descarga puede tardar varios minutos dependiendo de la conexión. Verifica que el modelo está disponible:

```bash .noeval
ollama list
```


### Inicio del servidor

Ollama ejecuta un servidor local que expone los modelos a través de una API REST. En la mayoría de instalaciones, el servidor se inicia automáticamente. Si necesitas iniciarlo manualmente:

```bash .noeval
ollama serve
```

El servidor escucha por defecto en `http://localhost:11434`. Puedes probar que funciona correctamente:

```bash .noeval
curl http://localhost:11434/api/tags
```

> Los modelos de Ollama se almacenan localmente y no requieren conexión a internet una vez descargados, lo que garantiza privacidad y funcionamiento offline.

## Configuración de Tavily Search

**Tavily** es un motor de búsqueda diseñado específicamente para agentes de IA. A diferencia de las APIs de búsqueda tradicionales, Tavily devuelve resultados estructurados y optimizados para ser consumidos por modelos de lenguaje.

### Obtención de la clave API

Para utilizar Tavily necesitas una clave API:

* **1.** Accede a [tavily.com](https://tavily.com) y crea una cuenta
* **2.** Navega al panel de control
* **3.** Copia tu clave API

El plan gratuito incluye un número limitado de búsquedas mensuales, suficiente para desarrollo y pruebas.

### Variables de entorno

Las claves API y credenciales sensibles deben almacenarse en variables de entorno. Crea un archivo `.env` en la raíz del proyecto:

```bash .noeval
TAVILY_API_KEY=tvly-xxxxxxxxxxxxxxxxxxxxx
```

Instala el paquete para cargar variables de entorno:

```bash .noeval
uv add python-dotenv
```

> El archivo `.env` debe añadirse a `.gitignore` para evitar que las credenciales se suban al repositorio.

### Carga de variables de entorno

Al inicio de cada script, carga las variables de entorno:

````
import os
from dotenv import load_dotenv

load_dotenv()

tavily_api_key = os.environ.get("TAVILY_API_KEY")
````

## Verificación del entorno completo

Antes de proceder a crear agentes, verifica que todas las dependencias están correctamente instaladas y configuradas. Crea un archivo `verificar_entorno.py`:

````
import os
from dotenv import load_dotenv

load_dotenv()

def verificar_langchain():
    import langchain
    print(f"LangChain: {langchain.__version__}")

def verificar_ollama():
    from langchain_ollama import ChatOllama
    llm = ChatOllama(model="gemma3:1b")
    response = llm.invoke("Responde solo 'OK'")
    print(f"Ollama gemma3:1b: {response.content.strip()}")

def verificar_tavily():
    from langchain_tavily import TavilySearch
    api_key = os.environ.get("TAVILY_API_KEY")
    if api_key:
        print("Tavily: API key configurada")
    else:
        print("Tavily: API key no encontrada")

if __name__ == "__main__":
    print("Verificando entorno...\n")
    
    verificar_langchain()
    verificar_ollama()
    verificar_tavily()
    
    print("\nEntorno verificado correctamente")
````

Ejecuta el script:

```bash .noeval
uv run verificar_entorno.py
```

Si todas las verificaciones pasan sin errores, el entorno está listo para comenzar a desarrollar agentes de IA.
