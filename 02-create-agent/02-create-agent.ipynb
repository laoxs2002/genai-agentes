```python
from langchain.agents import create_agent

agent = create_agent("gpt-5-nano") # modelo de OpenAI

inputs = {"messages": [{"role": "user", "content": "¿Cuánto es 1 + 1?"}]}

response = agent.invoke(inputs)
response
```

## Parámetros

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model="gpt-5-nano",
    temperature=0.1,
    max_tokens=1000,
    timeout=30
)
agent = create_agent(model)
```

## System prompt

```python
agent = create_agent(
    model,
    system_prompt="Solo puedes responder preguntas sobre Python",
)
```

## Entrada

```python
from langchain.agents import AgentState
from langchain.agents import create_agent

agent = create_agent("gpt-5-nano")

state = AgentState(messages=[{"role": "user", "content": "Me llamo Mike, ¿cuánto es 1 + 1?"}])
response = agent.invoke(state)
response
```

```python
for message in response["messages"]:
    message.pretty_print()
```

```python
# Conversación sin memoria:

agent = create_agent("gpt-5-nano")

inputs = {"messages": [{"role": "user", "content": "Me llamo Mike, ¿cuánto es 1 + 1?"}]}

response = agent.invoke(inputs)
response
```

```python
from langchain.messages import HumanMessage

response['messages'].append(HumanMessage('¿Cómo me llamo?'))
response = agent.invoke(response)
response
```

```python
# acceder al ultimo mensaje
print(response["messages"][-1].content)
```
