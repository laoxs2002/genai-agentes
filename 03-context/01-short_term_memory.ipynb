# Short-term memory

Short-term memory (thread-level persistence) enables agents to track multi-turn conversations. 

Checkpointer requires one or more of the following 'configurable' keys: thread_id, checkpoint_ns, checkpoint_id

## InMemorySaver

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent("gpt-5.1", checkpointer=InMemorySaver())

response = agent.invoke(
    {"messages": [{"role": "user", "content": "Hola me llamo Mike, ¿cuánto es 1 + 1 ?."}]},
    {"configurable": {"thread_id": "1"}},  
)
response
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "¿Cómo me llamo?."}]},
    {"configurable": {"thread_id": "1"}},  
)
response
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "¿Cuál fue la suma que te pedí?."}]},
    {"configurable": {"thread_id": "1"}},  
)
response
```

## SqliteSaver  

pip install langgraph-checkpoint-sqlite

```python
import sqlite3
from langgraph.checkpoint.sqlite import SqliteSaver

conn = sqlite3.connect("checkpoints.sqlite", check_same_thread=False)

checkpointer = SqliteSaver(conn)
checkpointer.setup()
```

```python
agent = create_agent("gpt-5.1", checkpointer=checkpointer)
agent.invoke(
    {"messages": [{"role": "user", "content": "Hola me llamo Mike, ¿cuánto es 1 + 1 ?."}]},
    {"configurable": {"thread_id": "2"}},  
)
```

```python
agent.invoke(
    {"messages": [{"role": "user", "content": "¿Cómo me llamo?."}]},
    {"configurable": {"thread_id": "2"}},  
)
```
